### Random Walk

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

First we will generate a list of random numbers and plot:

```python
series = np.random.rand(200)
plt.plot(series)
```

A **random walk** is different from a list of random numbers because the next value in the sequence is a modification of the previous value in the sequence. Here is simple model of a random walk:

1) Start from a given initial position.

2) Randomly select a -1 or 1 and add it to the observation from the previous time step.

3) Repeat step 2 for as long as you like (or until you achieve the desired number of steps).


In the cell below, define the variable `rand_walk`, a 1d numpy array with shape `(N,)` 
where N = 1000 points, using the algorithm above. The initial position is given in the PrairieLearn question page as `initial_position` (which is the first entry of the numpy array).
You may find np.random.choice() useful.

```python
N = 1000
```

```python
#grade (enter your code in this cell - DO NOT DELETE THIS LINE) 
initial_position = 10
rand_walk = [initial_position]
while len(rand_walk) < N:
    delta = np.random.choice([-1, 1])
    rand_walk.append(rand_walk[-1] + delta)
    
rand_walk = np.array(rand_walk)
```

```python
plt.plot(rand_walk)
```
